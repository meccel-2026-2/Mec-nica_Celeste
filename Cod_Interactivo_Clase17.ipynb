A partir de la ecuación $ r = \frac{p}{1 + e cos(\theta_{er})}$ vamos a crear un código interactivo. Cambiemos a coordenadas cartesianas con el eje x sobre el apoapsis y el eje y en el semilatus rectum

Trayectoria del vector relativo en el sistema perifocal 

```python
p = 1
e = 0.6
tetaers = np.linspace(0, 2 * np.pi, 1000)

rs = p / (1 + e * np.cos(tetaers))

xfs = rs * np.cos(tetaers)
yfs = rs * np.sin(tetaers)

plt.plot(xfs, yfs)
plt.axis('equal')
plt.grid()
plt.tight_layout()
```

```python
from ipywidgets import interact
```

```python
@interact(e = (0, 1, 0.1))
def dibuja_trayectoria(e=0.5):

    p = 1

    tetaers = np.linspace(0, 2 * np.pi, 1000)

    rs = p / (1 + e * np.cos(tetaers))

    xfs = rs * np.cos(tetaers)
    yfs = rs * np.sin(tetaers)
    
    plt.figure(figsize=(6, 6))
    plt.plot(xfs, yfs)
    
    Q = p / (1 - e)
    q = p / (1 + e)

    plt.xlim(- 2 * p, 2 * p)
    plt.ylim(- 2 * p, 2 * p)
    plt.grid()
    plt.tight_layout()
    plt.show()
```

Transformación de Jacobi

```python
m1 = 1
m2 = 1
M = m1 + m2

rvecs = np.array([xfs, yfs]).T

r1vecs = m2/ M * rvecs
r2vecs = - m1/ M * rvecs
```

Graficar las trayectorias

```python
plt.plot(r1vecs[:,0], r1vecs[:,1], label='cuerpo 1')
plt.plot(r2vecs[:,0], r2vecs[:,1], label='cuerpo 2')
plt.legend()
plt.axis('equal')
plt.legend()
plt.tight_layout()
plt.show()
```
