```python
import pymcel as pc
import numpy as np
import matplotlib.pyplot as plt
```

```python
e =	0.1911663355386932	
a =	0.9223803173917017	* pc.constantes.au
q =	0.7460522521429133	* pc.constantes.au	
i =	3.340958441017069 * np.pi / 180
node =	203.8996515621043 * np.pi / 180
argp =	126.6728325163065 * np.pi / 180	
peri =	126.6728325163065	
tp =	2461042.918242006079 * 86400 #Pasar a segundos
mu = pc.constantes.mu_sun # El mu de todo el sistema solar
```

Variables derivadas 

$$
\int_0^b \rho J_\nu(\chi_{\nu l} \rho / b) J_\nu(\chi_{\nu m} \rho / b) d\rho =
\begin{cases}
0 & \text{si } l \neq m \
\frac{b^2}{2} [J_{\nu+1}(\chi_{\nu l})]^2 & \text{si } l = m
\end{cases}
$$

```python
h = np.sqrt(a * mu * (1 - e **2))
b = a * np.sqrt(1 - e **2)
```

Escribir la ecuación de Kepler en la forma $ f(E) = 0$

```python
from astropy.time import Time
```

```python
t = Time("2029-04-13 18:52:00", scale = 'tdb').jd * 86400 #Pasar a días julianos y a segundos
t
```

```python
def ecuacion_kepler(E):
    funcion = E - e*np.sin(E) - h/(a*b) * (t - tp)
    return funcion
```

Resolver la ecuación

```python
from scipy.optimize import newton
```

```python
newton(ecuacion_kepler, 0)
```
